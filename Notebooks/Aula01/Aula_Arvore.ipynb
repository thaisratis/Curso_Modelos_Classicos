## Resultados a serem imprimidos

```python
def results(y_test, y_pred):
    """
    Exibe a matriz de confusão, a acurácia e o relatório de classificação para os valores previstos.

    Parâmetros:
    y_test (array-like): Valores reais das classes.
    y_pred (array-like): Valores previstos das classes.

    Retorna:
    None
    """
    # Calcula a matriz de confusão
    results = confusion_matrix(y_test, y_pred)
    print('Confusion Matrix :')
    print(results)
    
    # Calcula a acurácia
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy: %.2f%%" % (accuracy * 100.0))
    
    # Exibe o relatório de classificação
    print('Report :')
    print(classification_report(y_test, y_pred))
```

# Árvore de decisão

## Classificação

```python
def computeClassificationDecisionTree(X_train, X_test, y_train, y_test, printResults):
    """
    Treina um modelo de árvore de decisão para classificação e faz previsões nos dados de teste.

    Parâmetros:
    X_train (array-like): Conjunto de características de treinamento.
    X_test (array-like): Conjunto de características de teste.
    y_train (array-like): Conjunto de rótulos de treinamento.
    y_test (array-like): Conjunto de rótulos de teste.
    printResults (bool): Se True, imprime a matriz de confusão, a acurácia e o relatório de classificação.

    Retorna:
    tuple: Uma tupla contendo:
        - y_pred (array-like): Valores previstos das classes para o conjunto de teste.
        - arvore_classificacao (DecisionTreeClassifier): O modelo de árvore de decisão treinado.
    """
    # Cria o modelo de árvore de decisão
    arvore_classificacao = tree.DecisionTreeClassifier()
    
    # Treina o modelo
    arvore_classificacao.fit(X_train, y_train)
    
    # Faz previsões nos dados de teste
    y_pred = arvore_classificacao.predict(X_test)
    
    # Se printResults for True, imprime os resultados
    if printResults:
        results(y_test, y_pred)
    
    return y_pred, arvore_classificacao
```

```python
y_pred,arvore_classificacao = computeClassificationDecisionTree(X_train, X_test, y_train, y_test, True)
```

### Separando aleatoriamente a base de treino (80% da base original) e a base de teste (20% da base original)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
def results_regression(y_test, y_pred):
    """
    Calcula e exibe várias métricas de avaliação para um modelo de regressão.

    Parâmetros:
    y_test (array-like): Valores reais das variáveis dependentes.
    y_pred (array-like): Valores previstos das variáveis dependentes.

    Retorna:
    None
    """
    # Calcula o erro quadrático médio (MSE)
    mse = mean_squared_error(y_test, y_pred)
    print(f"mse: {mse}")
    
    # Calcula a raiz do erro quadrático médio (RMSE)
    rmse = math.sqrt(mse)
    print(f"rmse: {rmse}")
    
    # Calcula o erro absoluto médio (MAE)
    mae = mean_absolute_error(y_test, y_pred)
    print(f"mae: {mae}")
    
    # Calcula o erro percentual absoluto médio (MAPE)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    print(f"mape: {mape}")
    
    # Calcula o coeficiente de determinação (R²)
    r2 = r2_score(y_test, y_pred)
    print(f"r2_score: {r2}")
```

```python
def computeRegressionDecisionTree(X_train, X_test, y_train, y_test, printResults):
    """
    Treina um modelo de árvore de decisão para regressão e faz previsões nos dados de teste.

    Parâmetros:
    X_train (array-like): Conjunto de características de treinamento.
    X_test (array-like): Conjunto de características de teste.
    y_train (array-like): Conjunto de valores reais de treinamento.
    y_test (array-like): Conjunto de valores reais de teste.
    printResults (bool): Se True, imprime várias métricas de avaliação para o modelo de regressão.

    Retorna:
    tuple: Uma tupla contendo:
        - y_pred (array-like): Valores previstos para o conjunto de teste.
        - arvore_regressao (DecisionTreeRegressor): O modelo de árvore de decisão treinado.
    """
    # Cria e treina o modelo de árvore de decisão para regressão
    arvore_regressao = tree.DecisionTreeRegressor()
    arvore_regressao.fit(X_train, y_train)
    
    # Faz previsões nos dados de teste
    y_pred = arvore_regressao.predict(X_test)
    
    # Se printResults for True, imprime as métricas de avaliação
    if printResults:
        results_regression(y_test, y_pred)
    
    return y_pred, arvore_regressao
```

```python
y_pred,arvore_regressao = computeRegressionDecisionTree(X_train, X_test, y_train, y_test, True)
```
